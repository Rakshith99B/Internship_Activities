# Day21 - Intro to Machine Learning

## Types of Machine Learning

### Supervised Learning
Supervised Learning is a type of machine learning where the data contains labels or correct answers. The model learns from input data and corresponding outputs so it can predict results for new data. It is commonly used for classification and prediction tasks.

### Unsupervised Learning
Unsupervised Learning uses data without labels. The model tries to find hidden patterns, structures, or clusters in the data by identifying similarities and differences on its own.

### Reinforcement Learning
Reinforcement Learning is a learning approach where an agent learns by interacting with an environment and receiving rewards or penalties. The goal is to maximize rewards by improving decisions over time.
